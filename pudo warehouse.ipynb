```python
import pandas as pd
import numpy as np

ye_data = pd.read_parquet("dataset/yellow_tripdata_2023-10.parquet")
gr_data = pd.read_parquet("dataset/green_tripdata_2023-10.parquet")
fhvhv_data = pd.read_parquet("dataset/fhvhv_tripdata_2023-10.parquet")
```

```python
ye_pudo = ye_data[['tpep_pickup_datetime', 'tpep_dropoff_datetime', 'PULocationID', 'DOLocationID', 'trip_distance', 'total_amount']].rename(columns={
    'tpep_pickup_datetime': 'putime',
    'tpep_dropoff_datetime': 'dotime',
    'PULocationID': 'pickup_location_id',
    'DOLocationID': 'dropoff_location_id',
    'trip_distance': 'trip_distance',
    'total_amount': 'total_amount'
})

gr_pudo = gr_data[['lpep_pickup_datetime', 'lpep_dropoff_datetime', 'PULocationID', 'DOLocationID', 'trip_distance', 'total_amount']].rename(columns={
    'lpep_pickup_datetime': 'putime',
    'lpep_dropoff_datetime': 'dotime',
    'PULocationID': 'pickup_location_id',
    'DOLocationID': 'dropoff_location_id',
    'trip_distance': 'trip_distance',
    'total_amount': 'total_amount'
})

fhvhv_data['total_amount'] = fhvhv_data['base_passenger_fare'] + fhvhv_data['tolls'] + fhvhv_data['bcf'] + fhvhv_data['sales_tax'] + fhvhv_data['congestion_surcharge'] + fhvhv_data['airport_fee'] + fhvhv_data['tips']
fhvhv_pudo = fhvhv_data[['pickup_datetime', 'dropoff_datetime', 'PULocationID', 'DOLocationID', 'trip_miles', 'total_amount']].rename(columns={
    'pickup_datetime': 'putime',
    'dropoff_datetime': 'dotime',
    'PULocationID': 'pickup_location_id',
    'DOLocationID': 'dropoff_location_id',
    'trip_miles': 'trip_distance',
    'total_amount': 'total_amount'
})

pudo = pd.concat([ye_pudo, gr_pudo, fhvhv_pudo]).reset_index(drop=True)
```

```python
pudo['putimestamp'] = pudo['putime'].dt.time
pudo['pudate'] = pudo['putime'].dt.date
pudo['dotimestamp'] = pudo['dotime'].dt.time
pudo['dodate'] = pudo['dotime'].dt.date
pudo.head(20)
```

```python
pudo.info()
```

```python
pudo['trip_id'] = range(1, len(pudo) + 1)
pu = pudo[['trip_id', 'putime', 'putimestamp', 'pudate', 'pickup_location_id', 'trip_distance', 'total_amount']]
pu['type'] = 'pu'
do = pudo[['trip_id', 'dotime', 'dotimestamp', 'dodate', 'dropoff_location_id', 'trip_distance', 'total_amount']]
do['type'] = 'do'
```

```python
pu = pu.rename(columns={
    'trip_id': 'trip_id',
    'putime': 'time', 
    'putimestamp': 'timestamp', 
    'pudate': 'date',
    'pickup_location_id': 'location_id',
    'trip_distance': 'trip_distance',
    'total_amount': 'total_amount',
    'type': 'type'
})

do = do.rename(columns={
    'trip_id': 'trip_id',
    'dotime': 'time', 
    'dotimestamp': 'timestamp', 
    'dodate': 'date',
    'dropoff_location_id': 'location_id',
    'trip_distance': 'trip_distance',
    'total_amount': 'total_amount',
    'type': 'type'
})

pudo = pd.concat([pu, do], ignore_index=False)
```
